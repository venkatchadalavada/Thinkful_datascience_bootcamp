```
import pandas as pd
census_population = pd.read_csv("state-populations.csv")
census_regions = pd.read_csv("census-divisions.csv")

print(census_population.columns)
print(census_regions.columns)
```

```
# Merge these two DataFrames into a single DataFrame called census
census = census_population.merge(census_regions)

print(census.columns)
print(census.shape)
```

```
census
```

```
# Re-shape 'census' so that one column contains all population measures, and another year attributes

census_data = pd.melt(census,
                      id_vars = ['state', 'region', 'division'],
                      var_name = 'year',
                      value_name = 'population',
                      )

census_data
```

```
# Group the data by year and summarize it

census_data.groupby(['year']).describe()
```

```
# Group the data by region, division and year and summarize it (also included Population)

census_data.groupby(['region', 'division', 'year'])['population'].count()
```
