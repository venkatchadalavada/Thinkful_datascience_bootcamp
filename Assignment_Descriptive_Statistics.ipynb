```
# 7. Is there a statistically significant difference between the prices of
# houses that have an open porch versus houses that do not have an open porch? 
# How did you handle null values?

# First, let's see what data we have
projects_df[['openporchsf', 'saleprice']].info()
```

```
# There are no null values in the relevant columns.
# let us assume that a 0 in the openporchsf column means that there is no open porch

open_porches = projects_df[['openporchsf', 'saleprice']].iloc[(projects_df['openporchsf'] > 0).values] # open porch
open_porches.info()
```

```
not_open_porches = projects_df[['openporchsf', 'saleprice']].iloc[(projects_df['openporchsf'] == 0).values] # no open porch
not_open_porches.info()
```

```
stats.ttest_ind(open_porches['saleprice'], not_open_porches['saleprice'])
```
