```
import pandas as pd
shots_df = pd.read_json('https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-analytics-bootcamp/tweets.json')

def printDataFrame(shots_df):
  print('The shape of this DataFrame is {}'.format(shots_df.shape))
  print('The columns in this DataFrame are {}'. format(shots_df.columns))
  print('The info:')
  print(shots_df.info())

printDataFrame(shots_df)
shots_df.head(5)

```
